```python
# initial libraries import

# for basic operations
import numpy as np
import pandas as pd

# warnings
import warnings
warnings.filterwarnings('ignore')

# for data visualizations
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()
```

```python
# dataset load

df = pd.read_csv('github_data.csv', delimiter=',')
```

```python
# How meny repositories are duplicated
df[df[['repository','author']].duplicated()].count()[0]
```

```python
# drop duplicated
df = df.drop_duplicates(subset=['repository','author'])
```

```python
# 10 Repos with most start
df.sort_values('stars', ascending=False).head(10)
```

```python
# 10 Repos with most start
df.sort_values('commits', ascending=False).head(10)
```

```python
df_collections_sum = df.groupby(['collection']).agg('sum').reset_index()
```

```python
plt.figure(figsize=(17,6))
title = '10 nejoblíbenějších projektů na stránce Github'
ax = sns.barplot(y="stars",x='repository', data=df.sort_values('stars', ascending=False).head(10))
ax.set_title(title,fontsize=20)
ax.set_xlabel('Repozitář', fontsize=15)
ax.set_ylabel('Počet hvězd',fontsize=15)
ax.tick_params(labelsize=14)
ax.figure.savefig("img/plot01.png")
```

```python
plt.figure(figsize=(20,6))
title = '7 nejoblíbenějších kolekcí na stránce Github'
ax = sns.barplot(y="stars",x='collection', data=df_collections_sum.sort_values('stars', ascending=False).head(7))
ax.set_title(title,fontsize=20)
ax.set_xlabel('Název kolekce', fontsize=20)
ax.set_ylabel('Počet hvězd',fontsize=20)
ax.tick_params(labelsize=14)
ax.figure.savefig("img/plot02.png")
```

```python
df_repo_sum = df.groupby(['collection', 'repository','author']).agg('sum').reset_index()
```

```python
plt.figure(figsize=(20,6))
title = '10 projektů s největší základnou vývojářů'
ax = sns.barplot(x='repository',y='contributors', data=df_repo_sum.sort_values('contributors',ascending=False).head(10))
ax.set_title(title,fontsize=20)
ax.set_xlabel('Projekt', fontsize=20)
ax.set_ylabel('Počet contributorů',fontsize=20)
ax.tick_params(labelsize=14)
ax.figure.savefig("img/plot03.png")
```

```python
plt.figure(figsize=(20,6))
title = '10 projektů s největším počtem commitů'
ax = sns.barplot(x='repository',y='commits', data=df_repo_sum.sort_values('commits',ascending=False).head(10))
ax.set_title(title,fontsize=20)
ax.set_xlabel('Projekt', fontsize=20)
ax.set_ylabel('Počet commitů',fontsize=20)
ax.tick_params(labelsize=14)
ax.figure.savefig("img/plot04.png")
```

```python
plt.figure(figsize=(20,6))
title = '10 projektů s největším množství nahlášených bugů'
ax = sns.barplot(x='repository',y='issues_open', data=df_repo_sum.sort_values('issues_open',ascending=False).head(10))
ax.set_title(title,fontsize=20)
ax.set_xlabel('Projekt', fontsize=20)
ax.set_ylabel('Počet bugů',fontsize=20)
ax.tick_params(labelsize=14)
ax.figure.savefig("img/plot05.png")
```

```python
non_lang_cols = ['collection', 'repository','author','stars','forks','watchers', 'dependents', 'commits', 
                 'branches', 'packages', 'releases', 'contributors', 'issues_open', 'issues_closed', 
                 'pull_requests_open', 'pull_requests_closed']
```

```python
df_lang = df.drop(non_lang_cols,axis=1)
df_lang.head(5)
```

```python
most_used_lang = df_lang[df_lang > 0 ].count().sort_values(ascending=False)
```

```python
plt.figure(figsize=(20,6))
title = '10 nejpoužívanějších technologií v projektech'
ax = sns.barplot(x='index',y=0, data=pd.DataFrame(most_used_lang).reset_index().head(10))
ax.set_title(title,fontsize=20)
ax.set_xlabel('Technologie', fontsize=20)
ax.set_ylabel('Počet projektů',fontsize=20)
ax.tick_params(labelsize=15)
ax.figure.savefig("img/plot06.png")
```
